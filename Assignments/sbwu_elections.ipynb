```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
from sklearn.preprocessing import StandardScaler,LabelEncoder
```

```python
url= "https://raw.githubusercontent.com/balaji2v/Inceptez_Batch19/main/sbwu_elections.csv"
df = pd.read_csv(url)
df
```

```python
df.info()
```

```python
df.describe(include="all").T
```

```python
df.corr()
```

```python
 lb=LabelEncoder()
```

```python
df["Result"]= lb.fit_transform(df["Result"])
```

```python
df
```

```python
df.drop(["Store","Location","State","RTWState","VoteCountDate"],axis="columns",inplace=True)
```

```python
df
```

```python
x = df.iloc[:,:-1]
x
```

```python
y= df.iloc[:,-1]
y
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=3,stratify=y)
```

```python
scl = StandardScaler()
```

```python
scl.fit(x_train)
```

```python
x_train_scl= scl.transform(x_train)
```

```python
x_train_scl
```

```python
x_test_scl = scl.transform(x_test)
```

```python
x_test_scl
```

```python
y_train
```

```python
y_test
```

```python
knn= KNeighborsClassifier(n_neighbors=3)
```

```python
knn
```

```python
knn.fit(x_train_scl,y_train)
```

```python
y_pred_train = knn.predict(x_train_scl)
```

```python
y_pred_test = knn.predict(x_test_scl)
```

```python
acc_score_train = accuracy_score(y_pred_train, y_train)
```

```python
acc_score_test = accuracy_score(y_pred_test, y_test)
```

```python
acc_score_train
```

```python
acc_score_test
```

```python
confusion_matrix(y_pred_train, y_train)
```

```python
confusion_matrix(y_pred_test,y_test)
```

```python
print(classification_report(y_pred_train,y_train))
```

```python
print(classification_report(y_pred_test,y_test))
```
